# Introducción General

  Tener empleo aporta muchos beneficios importantes, incluyendo proporcionar una fuente de ingresos, la mejora de la inclusión social, satisfacer las aspiraciones personales, fortalecer la autoestima y desarrollar capacidades y competencias. En Estados Unidos, alrededor del 70% de la población en edad laboral (entre 15 y 64 años) tiene un empleo remunerado. Esta cifra es mayor que el promedio de empleo de la OCDE de 68%.

  Sin embargo, cada uno de los trabajos esta conformado por factores de riesgo, los cuales son realizados a través de una determinación de la prima de riesgos de trabajo que se deriva de la revisión anual de la siniestralidad relacionando los casos de riesgos laborales terminados entre el 1 de enero y el 31 de diciembre de cada año.

  Las tareas de alto riesgo son todas las actividades que por su naturaleza o lugar donde se realiza, implica la exposición o intensidad mayor a las normalmente presentes en la actividad rutinaria las cuales pueden causar accidentes laborales severos y, en muchas ocasiones, mortales.

  En esta base de datos la Administración de Seguridad y Salud Ocupacional, también conocida como OSHA, exige a los empleadores que informen de todas las lesiones graves relacionadas con el trabajo, definidas como amputación, hospitalización o pérdida de un ojo siendo que el requisito comenzó el 1 de enero de 2015 y se registraron datos hasta el 28 de febrero de 2017.

  Dado lo anterior, podemos conocer generalmente cuál es el comportamiento de nuestra información con la finalidad de realizar estrategias para la reduccion de accidentes laborales severos y la implementación de mayor cuidado y prevención sobre la salud de los trabajadores.



# Descripción de datos

* *Descripción de las columnas*

  * Part of Body: Parte del cuerpo (cadena)

  * Source: Procedencia (cadena)
  * Secondary Source: Segunda procedencia (cadena)
